```python
most_popular = data.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
data.set_index("release_date", inplace=True)
data.index=pd.to_datetime(data.index)
data.head()
```

```python
data[["artists"]].iloc[18]
```

```python
data["duration"]=data["duration_ms"].apply(lambda x: round(x/1000))
data.drop("duration_ms", inplace=True, axis=1)
data.duration
```

### Correlation HeatMap

```python
corr_df=data.drop(["key","mode","explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables", fontsize = 18)
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90, fontsize = 14)
```

```python
sample = data.sample(int(0.004*len(data)))
print(len(sample))
```

### Regression Plot between Loudness and Energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample, y = "loudness", x = "energy", color = "c").set(title="Loudness vs Energy Correlation")
plt.show()
```

### Regression Plot between Popularity and Acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample, y = "popularity", x = "acousticness", color = "c").set(title="Popularity vs Acousticness Correlation")
plt.show()
```

```python
data["dates"]=data.index.get_level_values("release_date")
data.dates=pd.to_datetime(data.dates)
years=data.dates.dt.year
```

```python
total_duration = data.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y=total_duration, ax=ax, errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)
plt.show()
```

```python
total_duration=data.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_duration, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=90)
plt.show()
```

### Different Genre Analysis

```python
genre = pd.read_csv(r"C:\Users\sunabh824_2\Desktop\SpotifyFeatures.csv")
```

```python
plt.title("Duration of the songs in Different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y="genre", x="duration_ms", data=genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
plt.show()
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,5))
famous = genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(y="genre", x = "popularity", data = famous)
plt.title("Top 5 Famous Genres", fontsize = 20)
plt.show()
```
