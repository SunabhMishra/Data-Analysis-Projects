### Final Code

```python
import os
def get_topic_page(topic_url):
    #download the page
    response = requests.get(topic_url)
    #check successful response
    if response.status_code != 200:
        raise Exception ('Failed to load page {}'.format(topic_url))
    # parse using beautiful soup
    topic_doc = BeautifulSoup(response.text, 'html.parser')
    return topic_doc

def get_repository_info(h3_tag, star_tags):
    # returns all the info about the repository
    a_tags = h3_tag.find_all('a')
    username = a_tags[0].text.strip()
    repository_name = a_tags[1].text.strip()
    repository_url = base_url + a_tags[1]['href']
    stars = parse_star_count(star_tags.text.strip())
    return username, repository_name, repository_url, stars
    
def get_topic_repos(topic_doc):
    #get the h3 tags containing repo title, url, username
    h3_selection_class = 'f3 color-fg-muted text-normal lh-condensed'
    repository_tags = topic_doc.find_all('h3', {'class': h3_selection_class})
    # get star tags
    star_tags = topic_doc.find_all('span', {'class': 'Counter js-social-count'})
   
    repository_dict = {
        'username': [],
        'repo_name': [],
        'stars': [],
        'repo_url': []
    }
    
     # get repo info
    for i in range(len(repository_tags)):
        repo_info = get_repository_info(repository_tags[i], star_tags[i])
        repository_dict['username'].append(repo_info[0])
        repository_dict['repo_name'].append(repo_info[1])
        repository_dict['stars'].append(repo_info[2])
        repository_dict['repo_url'].append(repo_info[3])
        
    return pd.DataFrame(repository_dict)

def scrape_topic(topic_url, path):
    if os.path.exists(path):
        print("The file {} already exists.Skipping...".format(path))
        return
    topic_df = get_topic_repos(get_topic_page(topic_url))
    topic_df.to_csv(path, index = None)
```
